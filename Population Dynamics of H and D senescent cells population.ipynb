# Population Dynamics of Helper and Deleterious senescent cells

```python
#Import Libraries
import numpy as np
from scipy.integrate import odeint
from pylab import *
```

## Fig 1(b) and Fig 2(a) Phenotypic Transition at Normal rate (pt=0.005)

```python
def model(x,t):
    H,D = x
    dH = rh*H*(1 - (H+b*D)/K) - pt*H
    dD = rd*D*(1 - (a*H+D)/K) + pt*H #- s*D/
    return dH, dD

#Parameters
rh = 0.05
rd = 0.01
K = 10**8
pt = 0.005

ini = [100,50]
a = 1.0
b = 1.0
s = 0.0001
g = 0.5
t = np.arange(0,1000,0.01)
z = odeint(model,ini,t)
plot(t,z[:,0],'g',lw=3, label='H-SnC')
plot(t,z[:,1],'r',lw=3, label='D-SnC')
legend(fontsize=15)
xlabel('Time',fontsize=15)
ylabel('Population of SnC cells',fontsize=15)
xticks(fontsize=15)
yticks(fontsize=15)

#savefig('HD_SnC_PD.png', dpi=600)
show()
```

## Phenotypic Transition

## Fig 2(b) Low transition rate (pt=0.002)

```python
def model_SASP(x,t,es):
    H,D = x
    dH = rh*H*(1 - (H+b*D)/K) - pt*H
    dD = rd*D*(1 - (a*H+D)/K) + pt*H #- s*D/(g+D)
    return dH, dD

#Parameters
rh = 0.05
rd = 0.01
K = 10**8
pt = 0.005

ini = [100,50]
a = 1.0
b = 1.0
s = 0.002
g = 0.5
t = np.arange(0,1000,0.01)
z = odeint(model,ini,t)
plot(t,z[:,0],'--g',lw=3, alpha=0.5)
plot(t,z[:,1],'--r',lw=3, alpha=0.5)
pt = 0.002
z = odeint(model_SASP,ini,t,args=(pt,))
plot(t,z[:,0],'g',lw=3, label='H-SnC')
plot(t,z[:,1],'r',lw=3, label='D-SnC')
legend(fontsize=15)
xlabel('Time',fontsize=15)
ylabel('Population of SnC cells',fontsize=15)
xticks(fontsize=15)
yticks(fontsize=15)

#savefig('HD_SnC_low_PT_0.002.png', dpi=600)
show()
```
